```python
import numpy as np 

arr = np.array([12,84,30,58,21,42,10])

print(arr > 10)
print(arr[arr < 50])
print(arr[(arr > 20) & (arr > 80)])
print(arr[(arr > 10) | (arr < 80)])

m = arr < 20
print(arr[m])

arr[arr > 20] = 0
print(arr)
```

```python
import numpy as np 

arr = np.array([10,20,30,40,50,60,70,80,90])

b = np.where(arr > 50)
print(b)

c = np.where(arr > 50 , 1 , 0)
print(c)
```

```python
#fancy indexing 
import numpy as np 

# 1D array
arr = np.array([7,30,10,52,78])

print(arr[[1,3,2]]) # mention specific indexing 
print(arr[[4,4,1]]) # repeating and changing order 

# 2D array 
a2 = np.array([
    [1,2,3],
    [75,92,37],
    [12,95,46]
])

print(a2[[0,2]]) # accessing only rows 
print(a2[:,[0,2]]) # accessing only columns
print(a2[[1,2],[0,1]]) # row-column pair selection (1,0) , (2,1)

# 3D array 
a3 = np.array([
    [
        [2,89,10],
        [87,84,46]
    ],
    [
        [12,38,52],
        [39,52,65]
    ]
])

print(a3[[0]]) # accessing specific block 
print(a3[[1,0]]) # accessing specific block
print(a3[[1,0],[0,1],[2,2]]) # accessing specific element 
```

```python
# Universal functions 

a1 = np.array([12,63,70])

a2 = np.array([
    [48,39,72],
    [87,54,53]
])

print(np.add(a1 , a2))
print(np.subtract(a2 , a1))
print(np.multiply(a2 , a1))
print(np.divide(a2 , a1))
print(np.multiply(np.add(a1 , 6) , 7))
```

```python
# Aggregate function 

import numpy as np 

a1 = np.array([10,20,30,40,50])

print(np.sum(a1))
print(np.mean(a1))
print(np.min(a1))
print(np.max(a1))
print(np.median(a1))
print(np.std(a1))
```

```python
# Axis concept 

a1 = np.array([
    [56,39,76],
    [36,67,37]
])

print(np.sum(a1 , axis=0)) # remove row - work on column
print(np.min(a1 , axis=1)) # remove column - work on row 
print(np.max(a1 , axis=0))
```
