# Seaborn
- Seaborn is a Python data visualization library 
- built on top of Matplotlib 
- that works directly with Pandas DataFrames.

- Matplotlib = low-level 
- Seaborn = high-level 

### Why Seaborn?

- Seaborn is used because:
- Automatic good-looking graphs
- Works directly with Pandas DataFrames
- Built-in statistical plot
- Less code, more meaning
- Handles categorical + numerical data easily

✅ Real-life use:

- Data analysis
- Machine learning EDA
- Reports & dashboards

### Seaborn vs Matplotlib

Feature	Matplotlib	Seaborn

-   Level	    Low-level	    High-level
-   Appearance	Basic	Beautiful by default
-   Data handling	Lists/arrays	Pandas DataFrames
-   Statistical plots	Manual	Built-in
-   Code length	More	Less

### Basic Plot Structure in Seaborn
📌 Seaborn works on this concept:

DataFrame  →  Columns  →  Graph

### General Syntax:
sns.plot_name(data=df, x="column1", y="column2")

✔ data → Pandas DataFrame
✔ x, y → column names
✔ Seaborn handles axes automatically

Seaborn has:

1. sns.lineplot()
2. sns.scatterplot()
3. sns.barplot()
4. sns.histplot()

But internally:
👉 Seaborn creates Matplotlib Figure & Axes automatically

📌 Meaning:

- You don’t manually create figure & axes
- Seaborn does it for you
- But still uses Matplotlib behind the scenes

# Distribution Plots (Seaborn)

📌 What are Distribution Plots?

Simple meaning:

Distribution plots show how values are spread in a dataset.

They help answer:

- Where are most values?
- Are values concentrated or spread out?
- Is data skewed?
- What is the shape of data?

Example:

Are most bills around ₹10–₹20 or ₹50–₹100?

### 1️⃣ Histogram — histplot()
📌 What is a Histogram?

A histogram shows:

- X-axis → data values
- Y-axis → frequency (count)
- Data is divided into bins (intervals).

🔍 Syntax:

sns.histplot(data, bins=, kde=)

Important Parameters:
Parameter	Meaning
bins	Number of intervals
kde	Show density curve
color	Bar color
stat	count / frequency / density

# 2️⃣ KDE Plot — kdeplot()
📌 What is KDE?

KDE = Kernel Density Estimation

👉 It shows a smooth curve representing data distribution.

Think of it as:

Histogram → smooth line version

🔍 Syntax:

sns.kdeplot(data, x=)

Important Parameters:
Parameter	Meaning
fill	Fill area under curve
bw_adjust	Smoothness
multiple	Layering

# 3️⃣ Combined Histogram + KDE
📌 Why combine?

Histogram → exact counts

KDE → smooth shape

Best of both worlds ✅

Example:

sns.histplot(data=df, x="total_bill", kde=True)


# 4️⃣ Rug Plot
📌 What is a Rug Plot?

A rug plot shows:

Each data point as a small line (tick) on the axis

👉 Shows exact data positions

🔍 Syntax:

sns.rugplot(data, x=)

- Each small tick = one bill value
- Dense ticks → many values

🔹 Histogram vs KDE vs Rug (Quick Comparison)
Plot	Shows
Histogram	Frequency
KDE	Probability density
Rug	Exact values
Hist + KDE	Count + shape

# Categorical Plots 
📌 What are Categorical Plots?

Simple meaning:

Categorical plots compare numerical values across categories.

Example questions they answer:

- Which day has the highest total bill?
- Do males or females tip more?
- How does data vary across categories?

👉 X-axis → Category
👉 Y-axis → Numeric value
