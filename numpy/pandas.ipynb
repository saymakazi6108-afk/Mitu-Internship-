## Column Selection

# Data Cleaning

#### What are Missing Values?

Missing values are data entries where no value is stored.
In Pandas, missing values are represented as NaN (Not a Number).

🔹 isnull()

- Detects missing values in a dataset
- Returns True for missing values and False for non-missing values
- Used for identification, not removal
Use case: Finding incomplete records

🔹 notnull()

- Opposite of isnull()
- Returns True for valid data and False for missing values
- Useful for filtering valid rows

🔹 dropna()

- Removes rows or columns containing missing values
- Can drop:
-Rows with any missing values
-Columns with missing values
-Rows where all values are missing
-Risk of data loss if used without care
Use case: Removing invalid or incomplete records

🔹 fillna()

- Replaces missing values with specified values
- Can use:
-Constant values (0, "Unknown")
-Mean, median, or mode
-Forward fill or backward fill methods
- Preserves dataset size
Use case: Replacing missing salary with average salary


```python
import pandas as pd 

data = {
    "Name" : ["azam","neha","ronit","ritesh","saniya",None],
    "Age" : [18,56,None,33,None,None],
    "Marks" : [88,45,None,None,67,None]
}

df = pd.DataFrame(data)
print(df)
df.isnull()
df.notnull()
df[df["Age"].notnull() & df["Marks"].notnull()]

df2 = pd.DataFrame(data)
df.dropna() #remove row which have NaN

df2.dropna(axis = 1) #remove column which have NaN 

df2.dropna(how="all") #remove row only if all value in a row are NaN

df2.dropna(subset=["Marks"]) # remove row if specific column has NaN 

# fillna()
df2.fillna("unavailable") # fill NaN place with constant 
df2.fillna(0)

```

🔹 duplicated()

- Identifies duplicate rows
- Marks duplicates as True
- Can check duplicates based on specific columns
- Use case: Detect repeated records

🔹 drop_duplicates()

- Removes duplicate rows
- Keeps the first occurrence by default
- Can keep last occurrence or remove all duplicates
- Use case: Removing repeated customer records

🔹 replace()

- Substitutes specific values with new values
- Can replace:
1. Single values
2. Multiple values
3. Lists or dictionaries of values
- Helps in standardizing data
- Use case: Converting "M" → "Male", "F" → "Female"

```python
import pandas as pd 

data = {
    "Name" : ["neha","neha","ronit","ronit","saniya","saniya"],
    "Age" : [18,56,33,33,None,None],
    "Marks" : [88,45,45,45,67,None]
}

df = pd.DataFrame(data)
print(df)

df.duplicated()

df.duplicated(subset=["Name"]) # detect duplicate based on column

df.duplicated(keep="first") # keep first duplicate occurance

# drop duplicate

df.drop_duplicates()

df.drop_duplicates(subset=["Name"])# remove duplicate based on column

df.drop_duplicates(keep="last") # keep last duplicate record

df["Name"].replace("Saniya","sAniya")

```

✅ sort_values()

🔹 Meaning
- Sorts data by column values.
- by default ascending 

🔹 Syntax
- df.sort_values(by="column_name")
- df.sort_values(by="column_name" , ascending = False)

✅ sort_index()

🔹 Meaning
- Sorts data by index, not by column values.

sort_values() → columns
sort_index() → index only

```python
import pandas as pd

data = {
    "name": ["Ayaan", "Sara", "Ayaan", "Zara", "Sara"],
    "department": ["IT", "HR", "IT", "HR", "IT"],
    "age": [23, 29, 23, 35, 29],
    "salary": [50000, 60000, 52000, 58000, 61000]
}

df = pd.DataFrame(data)
print(df)

df.sort_values(by = "salary") # single column 
df.sort_values(by = ["department","salary"]) # using multiple columns 
df.sort_values(by="age" , ascending=False) # descending format

df.sort_index(ascending=False) # index wise sorting 
```

```python
import pandas as pd

data = {
    "name": ["Ayaan", "Sara", "Ayaan", "Zara", "Sara"],
    "department": ["IT", "HR", "IT", "HR", "IT"],
    "age": [23, 29, 23, 35, 29],
    "salary": [50000, 60000, 52000, 58000, 61000]
}

df = pd.DataFrame(data)
print(df)

df["salary"] = df["salary"].apply(lambda x: x * 1.10) #Applies a function row-wise or column-wise.
print(df)

df["department"].map({"IT": "Tech", "HR": "Human Resources"}) #Used on Series only to map values.

```

🔹 GROUPBY OPERATIONS
✅ groupby()

🔹 Meaning

Split data into groups → apply aggregation → combine result.

```python
import pandas as pd

data = {
    "name": ["Ayaan", "Sara", "Ayaan", "Zara", "Sara"],
    "department": ["IT", "HR", "IT", "HR", "IT"],
    "age": [23, 29, 23, 35, 29],
    "salary": [50000, 60000, 52000, 58000, 61000]
}

df = pd.DataFrame(data)
print(df)

df.groupby("department")["salary"].mean() #single column ,single agg function
df.groupby("department")["age"].agg(["min","max","mean"]) #single column , multiple agg function 
df.groupby(["name","department"])["salary"].max() #multiple column , single agg function 
df.groupby(["name","department"])["age"].agg(["min","max","mean"]) # multiple column, multiple agg function 
```

🔹 Working with Columns & Rows

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Ayaan", "Sara", "Zara"],
    "department": ["IT", "HR", "IT"],
    "salary": [50000, 60000, 55000]
})

print(df)
print("\n")

#Adding new column
df["bonus"] = 5000 #direct assignment 
print(df)
print("\n")


df["total_salary"] = df["bonus"] + df["salary"] #calculation from columns 
print(df)
print("\n")

df["tax"] = df["total_salary"].apply(lambda x : x * 0.10) #applying functions 
print(df)
print("\n")

# droping columns 
df.drop("bonus" , axis=1 , inplace=True) #single column
print(df)
print("\n")

df.drop([0,1] ,axis=0 ,inplace=True) # droping multiple rows 
print(df)
print("\n")

df.drop(["total_salary","tax"] ,axis=1 ,inplace=True) # droping multiple rows 
print(df)
print("\n")


#Adding new row 
df.loc[0] = ["fatima","IT",80000] #adding row using loc[]
print(df)
print("\n")

#using concat()
new_row = pd.DataFrame({
    "name": ["Ali"],
    "department": ["IT"],
    "salary": [58000]
})

df = pd.concat([df, new_row], ignore_index=True)
print(df)
print("\n")

```

🔹 Combining & Merging Data

🔹 1. CONCATENATION — concat()
✅ Meaning

Stack DataFrames together (row-wise or column-wise).

📌 Think: just stick them together, no matching keys.

🔹 2. MERGING DATA — merge()
✅ Meaning

Combine DataFrames based on common column (key).

📌 SQL-style joins.

🔹 INNER JOIN
🔹 Meaning

Keep only matching rows in both DataFrames.

🔹 LEFT JOIN
🔹 Meaning

All rows from left DataFrame, match from right.

🔹 3. JOIN USING INDEX — join()
✅ Meaning

Join DataFrames using index instead of column.
- Default join = LEFT JOIN

```python
import pandas as pd

employees = pd.DataFrame({
    "emp_id": [1, 2, 3, 4],
    "name": ["Ayaan", "Sara", "Zara", "Ali"],
    "department": ["IT", "HR", "IT", "Finance"]
})

print(employees)

print("\n")
salaries = pd.DataFrame({
    "emp_id" : [1,2,4,5],
    "salary" : [54000,48000,90000,50000]
})
print(salaries)

#concat()
pd.concat([employees, employees]) #Row-wise concatenation (default)
pd.concat([employees, employees], ignore_index=True) #It ignore the index 
pd.concat([employees,salaries] , axis=1) #Matches by index, NOT by column values.

#merge()
pd.merge(employees,salaries,on="emp_id",how="inner") #Keep only matching rows in both DataFrames.
pd.merge(employees,salaries,on="emp_id",how="left") #All rows from left DataFrame, match from right.
pd.merge(employees,salaries,on="emp_id",how="right") #All rows from right DataFrame, match from left.
pd.merge(employees,salaries,on="emp_id",how="outer") #Keep all rows from both DataFrames

#JOIN USING INDEX — join()
employees_indexed = employees.set_index("emp_id")
salaries_indexed = salaries.set_index("emp_id")

employees_indexed.join(salaries_indexed) #outer join
employees_indexed.join(salaries_indexed, how="inner") #inner join 

```

🔹 1. DATETIME BASICS
✅ Why DateTime matters

Raw dates are usually strings, Pandas can’t extract year/month until you convert them.

✅ Convert to datetime — to_datetime()
🔹 Meaning

Converts string → datetime object.

🔹 2. DATE EXTRACTION (.dt accessor)

- Attribute	    Meaning

- year	    =    Extract year
- month	    =    Extract month
- day	     =   Extract day
- day_name()  = 	Monday, Tuesday…

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Ayaan Khan", "Sara Ali", "Zara Sheikh"],
    "join_date": ["2023-05-10", "2022-12-01", "2024-01-15"],
    "email": ["Ayaan@Gmail.com", "sara_ali@yahoo.com", "zara.sheikh@outlook.com"]
})

print(df)
print("\n")

#convert to datetime 
print(df.dtypes)
print("\n")

df["join_date"] = pd.to_datetime(df["join_date"])
print(df.dtypes)
print("\n")

#date extraction
df["year"] = df["join_date"].dt.year
print(df)
print("\n")

df["month"] = df["join_date"].dt.month
print(df)
print("\n")

df["date"] = df["join_date"].dt.day
print(df)
print("\n")

df["day"] = df["join_date"].dt.day_name()
print(df)
print("\n")

#string operation 
df["lower_case"] = df["email"].str.lower()
print(df)
print("\n")

df["upper_case"] = df["name"].str.upper()
print(df)
print("\n")

df["first_name"] = df["name"].str.split(" ").str[0]
print(df)
print("\n")
```
