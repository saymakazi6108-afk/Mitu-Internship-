- Matplotlib was created by John D. Hunter.
- Matplotlib is open source and we can use it freely.
- Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

### Multiple Line Plots

Used to compare two or more datasets on the same graph.

Rule

- Call plt.plot() multiple times before plt.show().

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]

y1 = [10,20,30,40,50,60]
y2 = [5,10,15,20,25,30]

plt.plot(x,y1 , color='red')
plt.plot(x,y2 ,color='black')
plt.show()
```

### Line Styles

Line style controls how the line looks.

Style	    Code
1. Solid	 →   '-'
2. Dashed	 →  '--'
3. Dotted	→ ':'
4. Dash-dot	 → '-. '

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]

y1 = [2,4,6,8,10]
y2 = [3,6,9,12,15]
y3 = [4,8,12,16,20]
y4 = [5,10,15,20,25]

plt.plot(x,y1 , linestyle='-')
plt.plot(x,y2 , linestyle='--')
plt.plot(x,y3 , linestyle=':')
plt.plot(x,y4 , linestyle='-.')
plt.show()
```

### Understanding Plot Structure (VERY IMPORTANT)

A Matplotlib plot has:

- Figure → the whole window (canvas)
- Axes → the actual plotting area (graph)

Figure
 └── Axes
      ├── X-axis
      └── Y-axis


- If you control the figure → size & layout
- If you control the axes → limits, ticks, grid

1️⃣ Figure Size

Controls width & height of the graph.

syntax 
- plt.figure(figsize=(width, height))

2️⃣ Axis Limits

Used to restrict the visible area of axes.

1. xlim()
- plt.xlim(min, max)

2. ylim()
- plt.ylim(min, max)

3️⃣ Grid

Grid improves readability of data.

syntax
- plt.grid()

4️⃣ Ticks

Ticks are numbers shown on axes.

1. xticks()

plt.xticks(list) - X-axis shows only given tick values

2. yticks()

plt.yticks(list) - Y-axis ticks at specific intervals

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.show()

plt.figure(figsize=(8, 4))

a = [1, 2, 3, 4]
b = [10, 20, 30, 40]

plt.plot(a,b)
plt.xlim(1,3)
plt.ylim(0,25)
plt.show()

plt.figure(figsize=(8, 4))

m = [1, 2, 3, 4]
n = [10, 20, 30, 40]

plt.plot(m,n)
plt.grid()
plt.show()

plt.figure(figsize=(8, 4))

p = [1, 2, 3, 4]
q = [10, 20, 30, 40]

plt.plot(p,q)
plt.xticks([1, 2, 3, 4])
plt.yticks([0, 10, 20, 30, 40])
plt.show()


```

5️⃣ Pie Chart
📌 Theory

Pie chart shows proportions / percentage of a whole.

✔ Used when total = 100%

📌 Function
plt.pie(values)

Pie chart attributes
Attribute	Meaning
labels	Names of slices
autopct	Percentage display
startangle	Rotate chart
explode	Highlight slice



Plot Type	Use Case
Line	Trend over time
Bar	Compare categories
Barh	Long category names
Histogram	Data distribution
Scatter	Relationship
Pie	Percentage / share


```python
import matplotlib.pyplot as plt

#bar
plt.figure(figsize=(8,4))

production = [1,2,3,4,5]
Sales = [12,13,51,25,11]

plt.bar(production,Sales)
plt.xlabel("Production")
plt.ylabel("Sales")
plt.title("Daily production , sale report")
plt.show()

#horizontal bar
plt.figure(figsize=(8,4))

production1 = [1,2,3,4,5]
Sales1 = [12,13,51,25,11]

plt.barh(production1,Sales1)
plt.xlabel("Production")
plt.ylabel("Sales")
plt.title("Daily production , sale report")
plt.show()

#histogram
plt.figure(figsize=(8,4))

marks = [45, 50, 60, 70, 80, 85, 90, 95, 100]

plt.hist(marks , bins=5)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Marks Distribution")
plt.show()

#scatter
plt.figure(figsize=(8,4))

hours = [1, 2, 3, 4, 5]
scores = [40, 50, 65, 70, 85]

plt.scatter(hours, scores)
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.title("Study vs Score")
plt.show()

#pie chart
sizes = [40, 30, 20, 10]
labels = ["Python", "Java", "C", "Others"]

plt.pie(sizes, labels=labels)
plt.title("Programming Language Usage")
plt.show()


```

### Subplots


Subplots allow you to display multiple graphs in a single figure.

Function
plt.subplot(rows, columns, index)

📌 Meaning
Parameter	Meaning
rows	Number of rows
columns	Number of columns
index	Position (starts from 1)

fig, ax = plt.subplots() # single subplot 


Multiple Subplots using plt.subplots()
📌 Function
fig, axes = plt.subplots(rows, columns)



Layout Adjustment – tight_layout()
📌 Theory

Sometimes plots overlap (titles, labels clash).
tight_layout() fixes spacing automatically.

📌 Function
plt.tight_layout()

Feature	subplot()	subplots()
Style	MATLAB-like	Object-oriented
Control	Limited	High
Return axes	❌ No	✔ Yes
Best for	Simple plots	Projects

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]

y1 = [10,20,30,40,50]
y2 = [50,40,30,20,10]

plt.subplot(1,2,2)
plt.plot(x , y1)
plt.title("increasing")

plt.subplot(1,2,1)
plt.plot(x , y2)
plt.title("Decreasing")

plt.show()

#single subplot
fig, ax = plt.subplots()

ax.plot(x, y1)
ax.set_title("Simple Plot")

plt.show()

#multiple subplot 
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]

fig, axes = plt.subplots(2, 2)

axes[0, 0].plot(x, [10, 20, 30, 40])
axes[0, 0].set_title("Line 1")

axes[0, 1].plot(x, [40, 30, 20, 10])
axes[0, 1].set_title("Line 2")

axes[1, 0].bar(x, [5, 10, 15, 20])
axes[1, 0].set_title("Bar")

axes[1, 1].scatter(x, [3, 6, 9, 12])
axes[1, 1].set_title("Scatter")

plt.show()

import matplotlib.pyplot as plt

x = [1,2,3,4,5]

y1 = [10,20,30,40,50]

fig ,axes = plt.subplots(2,2)

for ax in axes.flat :
    ax.plot(x,y1)
    ax.set_title("plot")

plt.tight_layout() #fixes spacing automatically.
plt.show()
```

### Matplotlib with pandas's dataframe

```python
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.DataFrame({
    "day" : [1,2,3,4,5],
    "sale_A" : [18,43,24,11,23],
    "sale_B" : [12,23,13,52,45]
})

plt.plot(df["day"],df["sale_A"],label="Sale A")
plt.plot(df["day"],df["sale_B"],label="Sale B")

plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Sales")
plt.legend()
plt.show()
```
