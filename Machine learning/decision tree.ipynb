# Entropy
Entropy is a measure of randomness or impurity in a dataset.

In Decision Trees:

👉 Entropy tells us how mixed the classes are in a node.

Entropy measures:

How much information is needed to classify a random sample.

- High entropy → more information needed
- Low entropy → less information needed

__For binary classification, entropy ranges from 0 (pure node) to 1 (maximum impurity).__

<img src="images/1.png" width="400" height="250">


Imagine a box of balls.

Case 1:

All balls are RED.

👉 No confusion
👉 Perfectly pure
👉 Entropy = 0

Case 2:

50% RED, 50% BLUE.

👉 Maximum confusion
👉 Very impure
👉 Entropy = 1 (for binary case)

Case 3:

90% RED, 10% BLUE.

👉 Slight impurity
👉 Entropy is low but not zero

- Entropy = Measure of impurity
- Pure node → Entropy = 0
- Equal class mix → Maximum entropy
- Used to compute Information Gain
- Introduced by Claude Shannon



# Gini index
The Gini Index (also called Gini Impurity) is a metric used in classification problems, especially in decision tree algorithms, to measure how impure or mixed a dataset is.

It is most commonly used in CART (Classification and Regression Trees), which is implemented in libraries like scikit-learn.

How likely is a randomly chosen data point to be incorrectly classified if we label it randomly according to the class distribution in that node?


- Low Gini value → Node is pure (mostly one class)
- High Gini value → Node is impure (mixed classes)

Formula of Gini Index

$$
Gini = 1 - \sum_{i=1}^{n} p_i^2
$$​

Where:

- n = number of classes
- 𝑝 = probability of class i in the dataset

Example Calculation
Suppose a node contains 10 samples:

6 belong to Class A

4 belong to Class B

Step 1: Calculate probabilities

- \( p(A) = \frac{6}{10} = 0.6 \)
- \( p(B) = \frac{4}{10} = 0.4 \)

Step 2: Apply formula

\[
\text{Gini} = 1 - (0.6^2 + 0.4^2)
\]

\[
= 1 - (0.36 + 0.16)
\]

\[
= 1 - 0.52
\]

\[
= 0.48
\]

👉 This means the node is **48% impure**.


## Gini Index Values Range

| Gini Value | Meaning |
|------------|---------|
| **0** | Pure node (only one class) |
| **0.5** | Maximum impurity (binary classes, evenly split) |
| **> 0.5** | Highly mixed (multi-class problems) |


## How Decision Trees Use Gini Index

When building a decision tree:

1. All possible splits are evaluated  
2. **Gini Index is calculated for each split**  
3. The split with the **lowest weighted Gini Index** is selected  

### Weighted Gini Formula

$$
\text{Gini}_{split} =
\sum
\left(
\frac{\text{samples in child}}
{\text{total samples}}
\right)
\times
\text{Gini(child)}
$$

## Gini Index vs Entropy

| Feature | Gini Index | Entropy |
|----------|------------|----------|
| Used in | CART | ID3 / C4.5 |
| Formula complexity | Simple | More complex (log) |
| Speed | Faster | Slower |
| Sensitivity | Less sensitive to class imbalance | More sensitive |

🔹 Entropy (Information Theory)

Comes from Claude Shannon’s Information Theory

Measures uncertainty / information required

Answers:

How much information is needed to classify the data correctly?

🔹 Gini Index (Probability Theory)

Measures misclassification probability

Answers:

How often will a randomly chosen element be classified incorrectly?

📌 Same goal, different interpretation

#### Example of Split Selection (Gini Index)

Assume:

Left child Gini = 0.2 (4 samples)

Right child Gini = 0.4 (6 samples)

Weighted Gini Formula
$$
\text{Gini}_{split} =
\frac{4}{10} \times 0.2 +
\frac{6}{10} \times 0.4
$$
Final Calculation
$$
= 0.08 + 0.24
$$
$$
= 0.32
$$
Final Result

The split with lower Gini (0.32) is better.

# 🌳 Pruning in Machine Learning (Detailed Explanation)

Pruning is a technique mainly used in decision tree–based models to reduce the size of a tree by removing parts that provide little or no predictive power.

Its main goal is to prevent overfitting, improve generalization, and make the model simpler and faster.

### 🔹 Why Pruning Is Needed

Decision trees can grow very deep and start memorizing training data instead of learning patterns.

1. Problems without pruning:

- Overfitting ❌
- Poor performance on new (unseen) data
- Complex & hard-to-interpret trees
- Higher computation cost

2. Pruning helps to:

- Reduce variance
- Improve accuracy on test data
- Increase interpretability
- Reduce training & prediction time

#### 🔹 Where Pruning Is Used

Pruning is mainly applied in:

- Decision Tree
- Random Forest
- Gradient Boosting (Used in CART)

Formula:


Cost(T) = Error(T) + α × Number of Leaves

- α controls tree complexity
- Larger α → smaller tree

🔸 Minimum Error Pruning

Removes nodes that minimally increase training error


example :

let's assume

Tree A → Error = 5, Leaves = 10

Tree B → Error = 7, Leaves = 4

α = 0.5

Cost-Complexity Formula:

Cost(T)=Error(T)+α×Number of Leaves

🔹 Calculate Cost for Tree A

Cost(A)=5+(0.5×10)

Cost(A)=5+5

Cost(A)=10

🔹 Calculate Cost for Tree B

Cost(B)=7+(0.5×4)

Cost(B)=7+2

Cost(B)=9

✅ Final Decision

Since:

Cost(B) < Cost(A)

✔ Tree B is selected
✔ Tree B is simpler (fewer leaves)
✔ This is pruning in action
